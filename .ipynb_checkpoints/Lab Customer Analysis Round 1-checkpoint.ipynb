```python
# Rearrange the columns in the dataframe as needed
```

```python
file1 = file1[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income','Monthly Premium Auto','Number of Open Complaints','Total Claim Amount','Policy Type','Vehicle Class']]
```

```python
# Concatenate the three dataframes
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data.shape
data
```

```python
# Which columns are numerical?
```

```python
data._get_numeric_data()
```

```python
# Which columns are categorical?
```

```python
data.select_dtypes('object')
```

```python
# Delete the column education and the number of open complaints from the dataframe.
```

```python
data = data.drop(['Education','Number of Open Complaints'], axis=1)
data.head()
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type
```

```python
data['Customer Lifetime Value'] = data['Customer Lifetime Value'] * 100
```

```python
data
```

```python
print(data['Customer Lifetime Value'].dtype)
```

```python
data['Customer Lifetime Value'] = data['Customer Lifetime Value'].astype(float)
```
