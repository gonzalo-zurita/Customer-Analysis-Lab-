```python
# Imputing the mean for the null values in numeric columns
```

```python
import numpy as np
```

```python
mean_months_since_last_claim = marketing_customer_analysis['Months Since Last Claim'].mean()
marketing_customer_analysis['Months Since Last Claim'] = marketing_customer_analysis['Months Since Last Claim'].fillna(mean_months_since_last_claim)
```

```python
marketing_customer_analysis.isna().sum()
```

```python
mean_number_of_open_complaints = marketing_customer_analysis['Number of Open Complaints'].mean()
marketing_customer_analysis['Number of Open Complaints'] = marketing_customer_analysis['Number of Open Complaints'].fillna(mean_number_of_open_complaints)
```

```python
marketing_customer_analysis.isna().sum()
```

```python
# Replacing null values for categorical variables
```

```python
marketing_customer_analysis['Response'].unique()
marketing_customer_analysis['Response'].value_counts()

marketing_customer_analysis['Response'].value_counts(dropna=False)
```

```python
marketing_customer_analysis['Response'] = marketing_customer_analysis['Response'].fillna('No')
```

```python
marketing_customer_analysis.isna().sum()
```

```python
marketing_customer_analysis['State'].value_counts()
```

```python
marketing_customer_analysis['State'] = marketing_customer_analysis['Response'].fillna('California')
```

```python
marketing_customer_analysis['Vehicle Class'].value_counts()
```

```python
marketing_customer_analysis['Vehicle Class'] = marketing_customer_analysis['Vehicle Class'].fillna('Four-Door Car')
```

```python
marketing_customer_analysis['State'] = marketing_customer_analysis['Response'].fillna('Medsize')
```

```python
marketing_customer_analysis['Vehicle Size'].value_counts()
```

```python
marketing_customer_analysis['Vehicle Size'] = marketing_customer_analysis['Vehicle Size'].fillna('Medsize')
```

```python
marketing_customer_analysis['Vehicle Type'].value_counts()
```

```python
marketing_customer_analysis = marketing_customer_analysis.drop(['Vehicle Type'], axis=1)
marketing_customer_analysis.head()
```

```python
marketing_customer_analysis.isna().sum()
```

```python
# Datetime format 
```

```python
marketing_customer_analysis["Effective To Date"]=pd.to_datetime(marketing_customer_analysis['Effective To Date'], errors='coerce')
```

```python
months=[]
for i in range(len(marketing_customer_analysis['Effective To Date'])):
    months.append(marketing_customer_analysis.iloc[i]['Effective To Date'].month)
marketing_customer_analysis['month']=months
```
