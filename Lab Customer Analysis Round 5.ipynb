```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

marketing_customer_analysis = pd.read_csv('marketing_customer_analysis.csv')
marketing_customer_analysis.head()
```

```python
marketing_data = marketing_customer_analysis.drop(['Unnamed: 0', 'unnamed: 0'], axis=1)
marketing_data.head()
```

```python
# X-y split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
y = marketing_data['customer lifetime value']
X = marketing_data.drop(['customer lifetime value'], axis=1)

# I choose the 'customer lifetime column', because the goal of this study 
# is to find what are the most profitable customers and 
# this variable seem the most representative for that matter
```

```python
X_num = X.select_dtypes(include = np.number)
```

```python
X_num
```

```python
# Normalizing 
```

```python
transformer = MinMaxScaler().fit(X_num)  
x_normalized = transformer.transform(X_num) 
pd.DataFrame(x_normalized, columns=X_num.columns)
```
